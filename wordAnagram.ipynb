```python
cleanword = sorted(list(set([words.strip().lower() for words in open('allwords','r')])))
cleanword[:40]
```

```python
def signature(word):
    return ''.join(sorted(word))

def anagram(myword):
    return [word for word in cleanword if signature(word) == signature(myword)]
```

```python
anagram('dictionary')
```

```python
anagram('lives')
```

```python
%timeit anagram('dictionary')
```

```python
%timeit anagram('lives')
```

```python
def sign(word):
    return ''.join(sorted(word))

def anag(myword):
    return [word for word in cleanword if sign(word)==sign(myword)]

anag('red')
```

```python
words_bysign = {}

for word in cleanword:
    words_bysign[sign(word)].append(word)
    
```

```python
import collections
```

```python
words_bysign = collections.defaultdict(list)

for word in cleanword:
    words_bysign[sign(word)].append(word)
```

```python
words_bysign
```

```python
def anagram_fast(myword):
    return words_bysign[sign(myword)]
```

```python
anagram_fast('Marry')
```

```python
anagram_fast('devil')
```

```python
%timeit anagram_fast('dictionary')
```
